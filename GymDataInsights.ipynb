```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/Users/muhamedhashi/Gym-Insights-Analysis/gym_members_exercise_tracking.csv' 
gym_data = pd.read_csv(file_path)

# Check for missing values and data types
missing_values = gym_data.isnull().sum()
data_types = gym_data.dtypes

# Descriptive Statistics
numeric_stats = gym_data.describe()
categorical_counts = gym_data['Workout_Type'].value_counts()

# BMI Insights
bmi_by_gender = gym_data.groupby('Gender')['BMI'].mean()
bmi_by_experience = gym_data.groupby('Experience_Level')['BMI'].mean()

# Correlation Analysis
correlation_matrix = gym_data[['Session_Duration (hours)', 'Max_BPM', 'Calories_Burned', 'Experience_Level']].corr()

# Water Intake and Performance
water_vs_calories_corr = gym_data['Water_Intake (liters)'].corr(gym_data['Calories_Burned'])

# Visualizations
# 1. Workout Frequency by Gender
plt.figure(figsize=(8, 6))
sns.barplot(data=gym_data, x='Gender', y='Workout_Frequency (days/week)', errorbar=None)
plt.title('Workout Frequency by Gender')
plt.ylabel('Workout Frequency (days/week)')
plt.xlabel('Gender')
plt.show()

# 2. Calories Burned by Workout Type
plt.figure(figsize=(10, 6))
sns.boxplot(data=gym_data, x='Workout_Type', y='Calories_Burned')
plt.title('Calories Burned by Workout Type')
plt.ylabel('Calories Burned')
plt.xlabel('Workout Type')
plt.xticks(rotation=45)
plt.show()

# 3. Session Duration vs. Calories Burned
plt.figure(figsize=(8, 6))
sns.scatterplot(data=gym_data, x='Session_Duration (hours)', y='Calories_Burned', hue='Workout_Type', edgecolor='w')
plt.title('Session Duration vs. Calories Burned')
plt.ylabel('Calories Burned')
plt.xlabel('Session Duration (hours)')
plt.legend(title='Workout Type')
plt.show()

# Print results
print("\n--- Missing Values and Data Types ---")
print(missing_values)
print(data_types)

print("\n--- Numeric Statistics ---")
print(numeric_stats)

print("\n--- Categorical Counts (Workout Type) ---")
print(categorical_counts)

print("\n--- BMI by Gender ---")
print(bmi_by_gender)

print("\n--- BMI by Experience Level ---")
print(bmi_by_experience)

print("\n--- Correlation Matrix (Key Metrics) ---")
print(correlation_matrix)

print("\n--- Correlation between Water Intake and Calories Burned ---")
print(water_vs_calories_corr)
```
